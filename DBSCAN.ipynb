### Import libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
import pandas as pd
from sklearn.cluster import DBSCAN
```

### Generate sample dataset (non-linear shape) using make_moons

```python
X,y=make_moons(n_samples=250,noise=0.05,random_state=42)
```

### Create a DataFrame for easier visualization and manipulation

```python
dataset = pd.DataFrame({
    "Feature 1": X[:, 0],
    "Feature 2": X[:, 1]
})
```

### Visualize the original dataset

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x="Feature 1", y="Feature 2", data=dataset, s=50)
plt.title("Original Data (make_moons)")
plt.show()
```

### Apply DBSCAN clustering

```python
dbscan = DBSCAN(eps=0.2, min_samples=5)
dataset["Cluster"] = dbscan.fit_predict(dataset[["Feature 1", "Feature 2"]])
```

 ### Visualize the clustered output

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(
    x="Feature 1",
    y="Feature 2",
    hue="Cluster",
    data=dataset,
    palette="tab10",
    s=60
)
plt.title("DBSCAN Clustering Results")
plt.legend(title="Cluster")
plt.show()
```
